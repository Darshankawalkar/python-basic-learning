load the training data trn and testing data from the csv files in the data directry separte the data matrix from the class variable


```python
#import the training data 
trn = pd.read_csv('optdigits_trn.csv')

#separate the attributes 
x_trn=trn.filter(regex='\d')
                 
#separte the class variable into y_turn
y_trn=trn.y
                 
#import the testing data                 
tst=pd.read_csv('optdigits_trn.csv')    
                 
#separte the attributes into x_test
X_test = tst.filter(regex='\d')
                 
#separte teh class variable into y_test
y_tst=tst.y
```

# PRINCIPAL COMPONENT ANANLYSIS:TRAIN MODEL
   in the training pahse we do as

1 setup the PCA Object

2 project the traing data onto the princial found by PCA

3 Plot The Ration of Variance explained by each direction

4 create a Scatter plot of objevt training Data along the first principal direction

5 Get the average log Likelihood score of all Smaples.


    

```python
#sets up the Pca project
pca=PCA()

#Transforms the training data('tf'='transformed')
trn_tf = pca.fit_transform(x_trn)

#plot the Variance explained by each component
plt.plot(pca.explained_variance_ratio_)
```

```python
#plot the projected data set on the first two pricipal components and colors by class
sns.scatterplot(
x=trn_tf[:,0],
y=trn_tf[:,1],

    style=y_trn,
hue=y_trn,
palette=['yellow', 'green', 'red'])
plt.title('Projected Data Set on the First Two Principal Components')
```

```python
# Plot the transformed data using the first two principal components and color it by class
plt.scatter(trn_tf[:, 0], trn_tf[:, 1], c=y_trn)
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('Projected Data Set on the First Two Principal Components')
plt.show()
```
